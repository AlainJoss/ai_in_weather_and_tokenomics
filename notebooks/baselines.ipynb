```python
from datetime import datetime, timedelta
import random

N = 300
start_date = "2021-01-01 00:00:00"
end_date   = "2023-01-01 00:00:00"

# parse strings
start_dt = datetime.strptime(start_date, "%Y-%m-%d %H:%M:%S")
end_dt   = datetime.strptime(end_date, "%Y-%m-%d %H:%M:%S")

# step size in hours (example: every 6h)
step = timedelta(hours=6)

# generate full range
dt_range = []
curr = start_dt
while curr <= end_dt:
    dt_range.append(curr)
    curr += step

# pick random indices
random_indices = random.sample(range(len(dt_range)), N)
sample_dts = [dt_range[i] for i in random_indices]

# optional ISO format
sample_dts = [dt.strftime("%Y-%m-%d %H:%M:%S") for dt in sample_dts]
sample_dts
```

```python
import torch
import numpy as np

from earth2studio.data import GFS
from earth2studio.data.utils import fetch_data
from earth2studio.utils.coords import map_coords
from earth2studio.utils.time import to_time_array
from earth2studio.models.px import FCN, SFNO
```

```python
models = {
    "fcn": FCN,
    "sfno": SFNO
}

for name, model in models.items():
    print(name, model)

    prognostic = model
    package = prognostic.load_default_package()
    prognostic = prognostic.load_model(package)

    prognostic_ic = prognostic.input_coords()

    if hasattr(prognostic, "interp_method"):
        interp_to = prognostic_ic
        interp_method = prognostic.interp_method
    else:
        interp_to = None
        interp_method = "nearest"

    x, coords = fetch_data(
        source=GFS(),
        time=to_time_array([sample_dts[0]]),
        variable=prognostic_ic["variable"],
        lead_time=prognostic_ic["lead_time"],
        device = torch.device("cpu"),
        interp_to=interp_to,
        interp_method=interp_method,
    )

    output_coords = prognostic.output_coords(prognostic_ic)

    x, coords = map_coords(x, coords , output_coords)
    shape = x.shape # shape: (len(variables), lon, lat)
    means, stds = np.zeros(shape), np.zeros(shape)  
    ex2 = np.zeros(shape)

    accepted_dates = []
    n = 0
    for date in sample_dts:
        time = to_time_array([date])
        try:
            x, coords = fetch_data(
                source=GFS(),
                time=time,
                variable=prognostic_ic["variable"],
                lead_time=prognostic_ic["lead_time"],
                device = torch.device("cpu"),
                interp_to=interp_to,
                interp_method=interp_method,
            )
            x, coords = map_coords(x, coords , output_coords)
        except FileNotFoundError:
            # missing S3 key (NoSuchKey) -> skip
            continue

        x = x.numpy().squeeze(0).squeeze(0)
        #path = name + "/" + date
        #np.save(path, x)

        means = (means * n + x) / (n + 1)
        ex2   = (ex2   * n + x**2) / (n + 1)

        accepted_dates.append(date)
        n+=1

    stds = np.sqrt(ex2 - means**2)
    np.save(f"{name}_stds", stds)
    np.save(f"{name}_means", means)
    with open(f'accepted_{str(model)}.txt', 'w') as f:
        for date in accepted_dates:
            f.write(f"{date}\n")
```
